 - **Test 5:** Check if age of outcome is less than age of intake for any animal. This would imply that the animal decreased in age in the shelter which is impossible.

```python
# Make an empty dataframe 
age_df = pd.DataFrame(
    columns=["number_intake","units_intake"])

#Split the column into 2 by the space
age_df[["number_intake", "units_intake"]] = df['AgeuponIntake'].str.split(" ", expand=True)

# Ues regex to replace the unit with corresponding number
age_df['units_intake'] = age_df["units_intake"].replace(regex={r'years?' : 365, r'days?' : 1, r'months?' : 30, r'weeks?' : 7})

# Change to float types
age_df['number_intake'] = age_df['number_intake'].astype('float64')
age_df['units_intake'] = age_df['units_intake'].astype('float64')

# Multiply the number by the units
age_df['AgeIntake_days'] = age_df['number_intake'] * age_df['units_intake']
```

```python
#Split the column into 2 by the space
age_df[["number_outcome", "unit_outcome"]] = df['AgeuponOutcome'].str.split(" ", expand=True)

# Ues regex to replace the unit with corresponding number
age_df['unit_outcome'] = age_df["unit_outcome"].replace(regex={r'years?' : 365, r'days?' : 1, r'months?' : 30, r'weeks?' : 7})

# Change to float types
age_df['number_outcome'] = age_df['number_outcome'].astype('float64')
age_df['unit_outcome'] = age_df['unit_outcome'].astype('float64')

# Multiply the number by the units
age_df['AgeOutcome_days'] = age_df['number_outcome'] * age_df['unit_outcome']
```

```python
age_difference = (age_df['AgeOutcome_days']-age_df['AgeIntake_days'])
test_5 = df[age_difference <0]
print("Number of rows failing the test: ", test_5.shape[0])
test_5.head(7)
```

```python
#Using code from the module lab

def make_bar_plots(data_f, col1, col2, title_of_graph):

    values = pd.unique(data_f[col1].ravel())

    # add new column and set values to zero
    data_f['percent'] = 0

    #print header
    print(col1)
    print("Index \t Count")

    # for each delinquency category
    for i in values:

        count = data_f[data_f[col1] == i].count()[col2]
        count_percentage = (1 / count) * 100

        # print out index vs count
        print(i, "\t", count)

        index_list = data_f[data_f[col1] == i].index.tolist()
        for ind in index_list:
            data_f.loc[ind, 'percent'] = count_percentage

    group = data_f[['percent',col1,col2]].groupby([col1,col2]).sum()

    my_plot = group.unstack().plot(kind='bar', stacked=True, title=title_of_graph, figsize=(15,7), grid=True)

    # add legend
    red_patch = mpatches.Patch(color='orange', label='Negative outcome')
    blue_patch = mpatches.Patch(color='blue', label='Positive outcome')
    my_plot.legend(handles=[red_patch, blue_patch], frameon = True)

    # add gridlines
    plt.grid(b=True, which='major', color='#666666', linestyle='-')
    plt.minorticks_on()
    
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)

    my_plot.set_xlabel(col1)
    my_plot.set_ylabel(col2)
    my_plot.set_ylim([0,100])
```
